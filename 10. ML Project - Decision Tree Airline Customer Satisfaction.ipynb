# 7) Create a function to display precision score, recall score, accuracy, classification report, confusion matrix, F1 Score.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

def evaluate_model(y_train, x_train_pred):

    accuracy = accuracy_score(y_train,x_train_pred)
    precision = precision_score(y_train, x_train_pred)
    recall = recall_score(y_train, x_train_pred)
    f1 = f1_score(y_train, x_train_pred)
    conf_matrix = confusion_matrix(y_train, x_train_pred)
    class_report = classification_report(y_train, x_train_pred)

    print(f"Accuracy Score: {accuracy:.2f}")
    print(f"Precision Score: {precision:.2f}")
    print(f"Recall Score: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
    print("\nConfusion Matrix:")
    print(conf_matrix)
    print("\nClassification Report:")
    print(class_report)
```

```python
evaluate_model(y_train, x_train_pred)
```
